#### Import Libraries and Load Data sets

```
# Import Libraries

%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import io
from datetime import datetime
```

```
# Dataset 1
# --------------------

# Load data file into pandas Dataframe
anscombe = pd.read_csv('anscombe.csv')
print(anscombe)

x1 = anscombe['x1']
y1 = anscombe['y1']

x2 = anscombe['x2']
y2 = anscombe['y2']

x3 = anscombe['x3']
y3 = anscombe['y3']

x4 = anscombe['x4']
y4 = anscombe['y4']

# Plot the data
plt.subplot(4, 1, 1)
plt.plot(x1, y1, 'o-')
plt.subplot(4, 1, 2)
plt.plot(x2, y2, 'o-')
plt.subplot(4, 1, 3)
plt.plot(x3, y3, 'o-')
plt.subplot(4, 1, 4)
plt.plot(x4, y4, 'o-')

# Show the plot
plt.show()
```
